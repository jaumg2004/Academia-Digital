### <font color='cyan'> Q20: Interprete o coeficiente de correlação, em termos de sinal e magnitude.</font>

Sinal positivo indica relação direta,quando uma variável aumenta, a outra tende a aumentar também, e quando uma diminui, a outra tende a diminuir. Magnitude 0,75 (aprox.): indica uma correlação forte / moderadamente forte
